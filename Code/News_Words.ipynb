# Feature Extraction

```python
import numpy as np
import pandas as pd
pd.options.display.max_columns = 999
```

```python
#loading the Boston Dataset
data = pd.read_csv('/Users/kyle/Documents/Virtual_Intern/Tencent/data/words.csv', low_memory=False)
df = pd.DataFrame(data)
```

```python
#Feature Engineering
# print(df_words[:10])
import string
def remove_punctuation(text):
    try: # python 2.x
        text = text.translate(None, string.punctuation)
    except: # python 3.x
        translator = text.maketrans('', '', '，？！《》.。、“”-+*/：……（）@:"＂,’‘— ')
        text = text.translate(translator)
    return text
```

```python
qid_without_punc = []
# Remove punctuation for qid
for i in range(len(df['qid'])):
    qid_without_punc.append(remove_punctuation(df['qid'][i]))
qid_without_punc = pd.DataFrame(qid_without_punc).rename(columns={0:"qid"})
qid_without_punc['qid'] = qid_without_punc['qid'].str.strip()
qid_without_punc = qid_without_punc['qid'].replace(' ', '')
df['qid_clean'] = qid_without_punc
df
```

```python
# Remove punctuation for title
title_without_punc = []
for i in range(len(df['title'])):
    title_without_punc.append(remove_punctuation((df['title'][i])))
title_without_punc = pd.DataFrame(title_without_punc).rename(columns={0:"title"})
title_without_punc['title'] = title_without_punc['title'].str.strip()
title_without_punc = title_without_punc['title'].replace(' ', '')
df['title_clean'] = title_without_punc
df
```

```python
df = df.drop(["qid",'title'], axis = 1)
df['qid_clean'] = df['qid_clean'].apply(lambda x: x.upper())
df['title_clean'] = df['title_clean'].apply(lambda x: x.upper())
df
```

```python
import jieba.analyse as anls
import jieba
def words_extract(sentence, cut_type):
    if cut_type == 'cut':
        tot_terms = []
        for i in range(len(sentence)):
            seg = jieba.lcut(sentence[i], cut_all = False)
            tot_terms.append(seg)
    elif cut_type == 'search':
        tot_terms = []
        for i in range(len(sentence)):
            seg = jieba.lcut_for_search(sentence[i])
            tot_terms.append(seg)
    return tot_terms
```

```python
df['qid_terms'] = words_extract(df['qid_clean'], cut_type = 'search')
df
```

```python
df['title_terms'] = words_extract(df['title_clean'], cut_type = 'search')
df
```

```python
df['num_qid_terms'] = df["qid_terms"].apply(lambda s: len(s))
df['num_title_terms'] = df["title_terms"].apply(lambda s: len(s))
df
```

```python
qid_cut_words = []
for i in range(len(df['qid_clean'])):
    cut_words = jieba.cut(df['qid_clean'][i], cut_all = False)
    for j in cut_words:
        qid_cut_words.append(j)
```

```python
qid_cut_words = list(set(qid_cut_words))
```

```python
qid_cut_words = pd.DataFrame(qid_cut_words).rename(columns={0:"qid"})
qid_cut_words
```

```python
important_words = []
for i in qid_cut_words['qid']:
    if len(i) >= 2:
        if len(i) <= 4:
            important_words.append(i)
len(important_words)
```

```python
important_words = [str(s) for s in important_words]
# Split out the words into individual columns
for word in important_words:
    df[word] = df['title_terms'].apply(lambda s : s.count(word))
```

```python
import re
import random
import string
letters = string.ascii_letters
df = df.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', random.choice(letters), x))
df
```

```python
def num_matches(terms_a, terms_b):
    match_density = 0
    times = 0
    for i in terms_a:
        for j in terms_b:
            if i == j:
                times += 1
    return times
```

```python
tot_num_matches = []
for i in range(len(df)):
    tot_num_matches.append(num_matches(df['qid_terms'][i],df['title_terms'][i]))
len(tot_num_matches)
```

```python
df['tot_num_matches'] = tot_num_matches
df[0:10]
```

```python
df['match_ratios'] = np.array(df['tot_num_matches']) * 1. / np.array(df['num_qid_terms'])
df[0:10]
```

# Model Set-up

```python
import lightgbm as lgb
from sklearn.datasets import load_breast_cancer, load_boston, load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import mean_squared_error, roc_auc_score, precision_score
```

```python
df_1 = data_after_extraction.drop(['label'], axis = 1)
Y_1 = data_after_extraction['label']
df_1
```

```python
#Scaling using the Standard Scaler
sc_1 = StandardScaler()
sc_1.fit(df_1)
X_1 = pd.DataFrame(sc_1.fit_transform(df_1))
```

```python
#train-test-split
X_train, X_test, y_train, y_test = train_test_split(df_1, Y_1, test_size=0.2, random_state=0)
```

```python
#Converting the dataset in proper LGB format
d_train = lgb.Dataset(X_train, label = y_train)
```

```python
#setting up the prarmeters
params = {}
params['learning_rate'] = 0.1
params['boosting_type'] = 'gbdt' # GradientBoostingDecisionTree
params['objective'] = 'multiclass' #Multi-class target feature
params['metric'] = 'multi_logloss' #metric for multi-class
params['max_depth'] = 40
params['num_leaves'] = 12
params['num_class'] = 6 #no.of unique values in the target class not inclusive of the end value
```

```python
# trainning the model
clf = lgb.train(params, d_train, 100) #targeting the model on 100 epocs
```

# Model Prediction

```python
#prediction on the test dataset
y_pred_1 = clf.predict(X_test)
```

```python
#printing the predctions
y_pred_1
```

```python
#argmax() method
y_pred_2 = [np.argmax(line) for line in y_pred_1]
#printing the predictions
print(y_pred_2[0:50])
```

```python
#using precision score for error metrics
prec_score = precision_score(y_pred_2, y_test, average=None).mean()
print(prec_score)
```

```python
#using RMSE error metric
RMSE = np.sqrt(mean_squared_error(y_pred_2, y_test))
print(RMSE) 
```
