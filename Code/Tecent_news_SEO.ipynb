```python
#Feature Engineering
import string
def remove_punctuation(text):
    try: # python 2.x
        text = text.translate(None, string.punctuation)
    except: # python 3.x
        translator = text.maketrans('', '', '，？！《》.。、“”-+*/：……（）@:"＂,’‘— ')
        text = text.translate(translator)
    return text
```

```python
# df = df.drop(['qid_clean', 'title_clean'], axis = 1)
# df
```

```python
df = df.drop(["qid",'title'], axis = 1)
df['qid_clean'] = df['qid_clean'].apply(lambda x: x.upper())
df['title_clean'] = df['title_clean'].apply(lambda x: x.upper())
# df
```

```python
import jieba.analyse as anls
import jieba
def words_extract(sentence, cut_type = 'cut'):
    if cut_type == 'cut':
        tot_terms = []
        for i in sentence:
            seg = jieba.lcut(i, cut_all = False)
            tot_terms.append(seg)
    elif cut_type == 'search':
        tot_terms = []
        for i in sentence:
            seg = jieba.lcut_for_search(i)
            tot_terms.append(seg)
    return tot_terms
```

```python
#Cut words in qid_clean
# df = df.drop(['qid_terms'], axis =1)
df['qid_terms'] = words_extract(df['qid_clean'], cut_type = 'cut')
df
```

```python
#Cut words in title_clean
df['title_terms'] = words_extract(df['title_clean'], cut_type = 'cut')
df
```

```python
#Compute the number of words cut for qid and title
df = df.drop(["qid_clean"], axis = 1)
df['num_qid_terms'] = df["qid_terms"].apply(lambda s: len(s))
df['num_title_terms'] = df["title_terms"].apply(lambda s: len(s))
```

```python
def num_matches(terms_a, terms_b):
    match_density = 0
    times = 0
    for i in terms_a:
        for j in terms_b:
            if i == j:
                times += 1
    return times
```

```python
tot_num_matches = []
for i in range(len(df)):
    tot_num_matches.append(num_matches(df['qid_terms'][i],df['title_terms'][i]))
len(tot_num_matches)
```

```python
#Feature 1: Number of matched words in title
df['tot_num_matches'] = tot_num_matches
# df[0:10]
```

```python
#Feature 2: The ratio between #of qid terms and #of meatched words in title
df['match_ratios'] = np.array(df['tot_num_matches']) * 1. / np.array(df['num_qid_terms'])
# df[0:10]
```

```python
#Feature 3: 
import jieba.analyse as anlys
def get_tfidf(list1, sentence):
    store = anlys.extract_tags(sentence, topK = None, withWeight= True, allowPOS=(), withFlag=False)
    TFIDF = []
    for i in store:
        TFIDF.append(i)
    words_with_tfidf = dict(TFIDF)
    
    words_weight = []
    for i in list1:
        for j in words_with_tfidf.keys():
            if i == j:
                words_weight.append(words_with_tfidf[j])
    return sum(words_weight)
```

```python
link = []
for i in range(len(df)):
    link.append(get_tfidf(df['qid_terms'][i], df['title_clean'][i]))
df['tfidf_score'] = link

```

```python
# Feature: match density
def match_density(list1, list2, density_threthold = 1):
    match_index = []
    count = 0
    for j in list2 :
        for i in list1:
            if j == i:
                match_index.append(list2.index(i))
                count += 1
    match_density = []
    for i in range(len(match_index)):
        try:
            if match_index[i+1] - match_index[i] <= density_threthold:
                match_density.append(1)
            else:
                match_density.append(0)
        except:
            break
    if match_density != [] :
    # return max(set(match_density), key=match_density.count)  # get mode
        return sum(match_density) * 1. / len(match_density)
    else :
        return 0
```

```python
link_match_density = []
for i in range(len(df)):
    link_match_density.append(match_density(df['qid_terms'][i], df['title_terms'][i], 2))
df['match_density'] = link_match_density
```

```python
# Feature: matched orders
def match_order_score(list1, list2):
    words_matched = []
    for j in list2:
        for i in list1:
            if j == i:
                words_matched.append(j)
#     print(words_matched)
    checkpoint1 = 0
    checkpoint2 = []
    from itertools import combinations, product
    checkpoint1 = list(combinations(list1, 2))
    for i in range(len(words_matched)):
        try:
            checkpoint2.append(tuple([words_matched[i], words_matched[i + 1]]))
        except:
            break

    matched_order = []
    for s in checkpoint1:
        for l in checkpoint2:
            if s == l:
                matched_order.append(1)
            else:
                matched_order.append(0)
    if matched_order == []:
        return 0
    else:
        return (matched_order.count(1) * 1. / len(matched_order))
```

```python
link_match_order = []
for i in range(len(df)):
    link_match_order.append(match_order_score(df['qid_terms'][i], df['title_terms'][i]))
df['match_order_score'] = link_match_order
```

# Model Set-up

```python
import lightgbm as lgb
from sklearn.datasets import load_breast_cancer, load_boston, load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import mean_squared_error, roc_auc_score, precision_score
```

```python
df_1 = data_after_extraction.drop(['label_binary', 'group_id'], axis = 1)
Y_1 = data_after_extraction['label_binary']
df_1, Y_1
```

```python
# #Scaling using the Standard Scaler
sc_1 = StandardScaler()
sc_1.fit(df_1)
X_1 = pd.DataFrame(sc_1.fit_transform(df_1))
X_1
```

```python
#train-test-split
X_train, X_test, y_train, y_test = train_test_split(df_1, Y_1, test_size=0.2, random_state=0)
```

```python
#Converting the dataset in proper LGB format
d_train = lgb.Dataset(X_train, label = y_train)
```

```python
#setting up the prarmeters
params = {}
params['learning_rate'] = 0.1
params['boosting_type'] = 'gbdt' # GradientBoostingDecisionTree
params['objective'] = 'binary' #Multi-class target feature
params['metric'] = 'binary_logloss' #metric for multi-class
params['max_depth'] = 30
params['num_leaves'] = 16 #no.of unique values in the target class not inclusive of the end value
```

```python
# trainning the model
clf = lgb.train(params, d_train, 100) #targeting the model on 100 epocs
```

```python
#prediction on the test dataset
y_pred_1 = clf.predict(X_test)
```
